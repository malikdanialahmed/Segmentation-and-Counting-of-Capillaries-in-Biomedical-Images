```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import frangi
from skimage.morphology import remove_small_objects, closing, square

# Path to your N1_Native folder
folder_path = 'Dataset/N1_Native'

# List all image files in the folder (assumes .png or .jpg)
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.tif'))]

def segment_capillaries(img):
    # Convert to grayscale if image is colored
    if len(img.shape) == 3:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        img_gray = img
    
    # Contrast stretching
    p2, p98 = np.percentile(img_gray, (2, 98))
    img_rescale = np.clip((img_gray - p2) / (p98 - p2), 0, 1)
    
    # Apply Frangi filter to enhance vessels
    vessels = frangi(img_rescale)
    
    # Show Frangi filter output for debug
    plt.imshow(vessels, cmap='gray')
    plt.title("Frangi vesselness output")
    plt.axis('off')
    plt.show()
    
    # Threshold the vesselness image - lower threshold
    thresh = vessels > 0.10
    
    # Morphological closing to fill gaps
    closed = closing(thresh, square(3))
    
    # Remove small objects (noise)
    cleaned = remove_small_objects(closed, min_size=30)
    
    return cleaned

# Process and display results for each image
for img_name in image_files:
    img_path = os.path.join(folder_path, img_name)
    img = cv2.imread(img_path)
    
    segmented = segment_capillaries(img)
    
    # Plot original and segmented side by side
    plt.figure(figsize=(10,5))
    plt.suptitle(f"Segmentation: {img_name}")
    
    plt.subplot(1,2,1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis('off')
    
    plt.subplot(1,2,2)
    plt.imshow(segmented, cmap='gray')
    plt.title("Segmented Capillaries")
    plt.axis('off')
    
    plt.show()
```

# Alternative 1: Adaptive Thresholding + Morphological Processing

```python
import cv2
import numpy as np
from skimage.morphology import remove_small_objects, closing, footprint_rectangle
from scipy.ndimage import binary_fill_holes
import matplotlib.pyplot as plt
import os

# Folder path where images are stored
dataset_path = "Dataset/N1_Native"

# List of images to process
images_to_process = ["N1a_Natif.jpg", "N1b_Natif.jpg"]  # Adjust extensions if needed

def load_image(image_name):
    path = os.path.join(dataset_path, image_name)
    img = cv2.imread(path)
    if img is None:
        raise FileNotFoundError(f"Image {image_name} not found at {path}")
    return img

# Adaptive Thresholding + Morphological Processing
def segment_capillaries_adaptive(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(img_gray, (5,5), 0)
    thresh = cv2.adaptiveThreshold(
        blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
        cv2.THRESH_BINARY_INV, 15, 10
    )
    binary = thresh.astype(bool)
    closed = closing(binary, footprint_rectangle(3, 3))
    cleaned = remove_small_objects(closed, min_size=50)
    return cleaned

# Canny Edge Detection + Morphology
def segment_capillaries_canny(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(img_gray, (3,3), 0)
    edges = cv2.Canny(blur, threshold1=30, threshold2=90)
    edges_bool = edges.astype(bool)
    closed = closing(edges_bool, footprint_rectangle(3, 3))
    filled = binary_fill_holes(closed)
    cleaned = remove_small_objects(filled, min_size=50)
    return cleaned

for image_name in images_to_process:
    img = load_image(image_name)
    segmented_adaptive = segment_capillaries_adaptive(img)
    segmented_canny = segment_capillaries_canny(img)

    plt.figure(figsize=(16,5))
    plt.suptitle(f"Segmentation Results: {image_name}", fontsize=16)

    plt.subplot(1,4,1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1,4,2)
    plt.title("Adaptive Thresholding")
    plt.imshow(segmented_adaptive, cmap='gray')
    plt.axis('off')

    plt.subplot(1,4,3)
    plt.title("Canny Edges")
    # Show raw edges
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(img_gray, (3,3), 0)
    edges = cv2.Canny(blur, threshold1=30, threshold2=90)
    plt.imshow(edges, cmap='gray')
    plt.axis('off')

    plt.subplot(1,4,4)
    plt.title("Canny + Morphology")
    plt.imshow(segmented_canny, cmap='gray')
    plt.axis('off')

    plt.show()
```
